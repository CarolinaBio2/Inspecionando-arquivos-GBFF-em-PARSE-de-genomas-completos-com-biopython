```python
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO
from Bio.SeqFeature import SeqFeature, FeatureLocation
```

```python
#ABRINDO O ARQUIVO E LENDO A FICHA DO GENOMA - ANOTAÇÕES GERAIS
#CDS - REGIÃO CODIFICADORA DO DNA
record = SeqIO.read("completo_genomic_ecoli.gbff", "genbank")
print(record)
print(record.annotations)
print(record.annotations['organism'])
```

```python
for record in SeqIO.parse("completo_genomic_ecoli.gbff", "genbank"):
    for feature in record.features:
      if feature.type == 'CDS':
        print(feature)  #SERVE COMO DEFINIÇÃO E EXEMPLO DO QUE É A FEATURE
```

```python
for record in SeqIO.parse("completo_genomic_ecoli.gbff", "genbank"):
    for feature in record.features:
      if feature.type == 'CDS':
        if 'gene' in feature.qualifiers:
          print(feature.qualifiers['gene'])
        else:
          print(list(feature.qualifiers))
```

```python
for record in SeqIO.parse("completo_genomic_ecoli.gbff", "genbank"):
  for feature in record.features:    #A FITA SENSO É 1 (+), E A FITA ANTISSENSO É -1 (-)
    print(record.features[2].location.start)  #ACESSANDO INÍCIO E FIM DE CADA FITA
    print(record.features[2].location.end)
```
